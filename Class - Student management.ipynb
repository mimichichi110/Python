```python
#Định nghĩa lớp sinh viên Student gồm các thông tin: mã sinh viên, họ tên, ngày sinh, quê quán.
# Phương thức info để in thông tin của sinh viên ra màn hình. Định nghĩa lớp danh sách sinh viên ListStudent
#thuộc tính là một list Student vừa định nghĩa ở trên.
#Các phương thức thêm sinh viên mới, cập nhật thông tin sinh viên theo mã sinh viên, xoá sinh viên trong list,
#tìm kiếm thông tin theo mã, in toàn bộ danh sách sinh viên ra màn hình.

class Student:
    def __init__(self, code, fullname, dob, residence):
        self.code = code
        self.fullname = fullname
        self.dob = dob
        self.residence = residence
    
    def info(self):
        print("{0}\t\t{1}\t\t{2}\t\t{3}".format(self.code, self.fullname, self.dob, self.residence))
```

```python
class listStudent:
    def __init__(self, _list):
        self._list = _list
        
    def initialize(self):
        stu1 = Student("1111", "Nguyễn Văn A", "22/10/1999", "Hà Nội")
        stu2 = Student("1112", "Nguyễn Văn B", "22/10/1996", "Hà Nội")
        stu3 = Student("1113", "Nguyễn C", "10/10/1999", "Hà Nội")
        stu4 = Student("1114", "Nguyễn Văn D", "22/10/2000", "Hà Nội")
        stu5 = Student("1115", "Nguyễn Văn CŨ", "22/10/1999", "HCM")
        
        self._list = [stu1, stu2, stu3, stu4, stu5]
    
    def output(self):
        print("Mã SV\t\tHọ và tên\t\tNgày sinh\t\tQuê quán")
        for i in self._list:
            i.info()
            
    def add(self, std):
        self._list.append(std)
        
    def update(self, code, std):
        '''code: mã sv sẽ bị chỉnh sửa
        std: thông tin sv sẽ được cập nhật'''
        for i in self._list:
            if i.code == code:
                i.fullname = std.fullname
                i.dob = std.dob
                i.residence = std.residence
                
    def delete(self, code):
        for i in self._list:
            if i.code == code:
                self._list.remove(i)
                break
    
    def search(self, code):
        for i in self._list:
            if i.code == code:
                return i
        
```

```python
class StudentManagement:
    def menu(self):
        print("==== Quản lý danh sách sinh viên ====")
        print("= 0. Load danh mục sinh viên       ==")
        print("= 1. Hiển thị danh mục sinh viên   ==")
        print("= 2. Thêm sinh viên                ==")
        print("= 3. Cập nhật sinh viên            ==")
        print("= 4. Xóa sinh viên                 ==")
        print("= 5. Tìm kiếm sinh viên            ==")
        print("= 6. Thoát                         ==")
        print("=====================================")
        
    def run(self):
        while True:
            self.menu()
            action = input("Nhập vào lựa chọn: ")
            if action == "0":
                self.list_st = listStudent(None)
                self.list_st.initialize()
                print("Load sinh viên thành công")
                input("Bấm enter để tiếp tục")
            elif action == "1":
                if self.list_st == None:
                    print("Bạn chưa load sinh viên")
                else:
                    self.list_st.output()
                input("Bấm enter để tiếp tục")
            elif action == "2":
                code = input("Nhập mã sinh viên: ")
                fullname = input("Nhập tên sinh viên: ")
                dob = input("Nhập ngày sinh: ")
                residence = input("Nhập quê quán: ")
                adding_student = Student(code, fullname, dob, residence)
                self.list_st.add(adding_student)
                print("Thêm sinh viên thành công")
                input("Bấm enter để tiếp tục")
            elif action == "3":
                code = input("Nhập mã sinh viên cần cập nhật: ")
                fullname = input("Nhập tên sinh viên: ")
                dob = input("Nhập ngày sinh: ")
                residence = input("Nhập quê quán: ")
                update_student = Student(code, fullname, dob, residence)
                self.list_st.update(update_student.code, update_student)
                input("Bấm enter để tiếp tục")
            elif action == "4":
                code = input("Nhập mã sinh viên cần xóa: ")
                self.list_st.delete(code)
                print("Xóa sinh viên thành công")
                input("Bấm enter để tiếp tục")
            elif action == "5":
                code = input("Nhập mã sinh viên cần tìm: ")
                self.list_st.search(code)
                input("Bấm enter để tiếp tục")
            elif action == "6":
                break
        print("Chương trình kết thúc")
```

```python
sm = StudentManagement()
sm.run()
```
