```python
import turtle

tl = turtle.Turtle()
tl.shape('turtle')
turtle.bgcolor('#87ceeb')

tl.penup()
tl.backward(250)
tl.pendown()

#House
tl.fillcolor('#1c00ff')
tl.begin_fill()
i = 0
while i < 2:
    tl.forward(200)
    tl.right(90)
    tl.forward(250)
    tl.right(90)
    i += 1
tl.end_fill()

#Door
tl.penup()
tl.goto(-110,-250)
tl.left(90)
tl.pendown()
tl.fillcolor('#90ef8f')
tl.begin_fill()
tl.forward(150)
tl.left(90)
tl.forward(80)
tl.left(90)
tl.forward(150)
tl.end_fill()

#Roof
tl.penup()
tl.goto(-250,0)
tl.left(135)
tl.pendown()
tl.fillcolor('#ff00ff')
tl.begin_fill()
tl.forward(141)
tl.right(90)
tl.forward(141)
tl.end_fill()

tl.fillcolor('#ffa500')
tl.begin_fill()
tl.left(60)
tl.forward(180)
tl.left(120)
tl.forward(141)
tl.left(60)
tl.forward(180)
tl.end_fill()

tl.penup()
tl.goto(-50, -250)
tl.right(180)
tl.pendown()
tl.fillcolor('yellow')
tl.begin_fill()
tl.forward(180)
tl.left(75)
tl.forward(250)
tl.left(105)
tl.forward(180)
tl.end_fill()
 
turtle.done()
```
